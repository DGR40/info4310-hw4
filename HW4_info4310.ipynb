```python
import pandas as pd
```

```python
drugs = pd.read_csv("drugs_unclean.csv")
```

```python
drugs2 = drugs.copy();
```

```python
drugs2 = drugs2.groupby(['therapeutic_area'])['therapeutic_area'].count().reset_index(name="total_count")
```

```python
drugs2= drugs2.sort_values(by=['total_count'], ascending=False)
```

```python
drugs3 = drugs2.head(10)
```

```python
drugs3
```

```python
goodlist = drugs3['therapeutic_area'].values.tolist()
```

```python
print(goodlist)
```

```python
drugs = drugs[drugs['therapeutic_area'].isin(goodlist)]
```

```python
drugs
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv') 

drugs.to_csv('out.zip', index=False,
          compression=compression_opts)  
```
